Removing the outliers and creating new data sets with it

There is a mathematical way to remove outliers  
IQR = Q3 - Q1  
Upper outlier = Q3 + 1.5 * IQR  
Lower outlier = Q1 - 1.5 * IQR  
These are weak outliers, so it may be dangerous to remove them all. Strong outliers are calculated by multiplying the IQR by 3 instead of 1.5. I will create two data sets, one with the strong outliers removed, and one with the weak outliers removed, and see the results of both.

```
crimeSC15bWeak <- crimeSC15a[which(crimeSC15a$NUM_CRIMES >= 513), ]
crimeSC16bWeak <- crimeSC16a[which(crimeSC16a$NUM_CRIMES >= 513), ]
crimeSC17bWeak <- crimeSC17a[which(crimeSC17a$NUM_CRIMES >= 513), ]
crimeSC18bWeak <- crimeSC18a[which(crimeSC18a$NUM_CRIMES >= 513), ]
crimeSC19bWeak <- crimeSC19a[which(crimeSC19a$NUM_CRIMES >= 513), ]
crimeSC15cWeak <- crimeSC15bWeak[which(crimeSC15bWeak$NUM_CRIMES <= 953), ]
crimeSC16cWeak <- crimeSC16bWeak[which(crimeSC16bWeak$NUM_CRIMES <= 953), ]
crimeSC17cWeak <- crimeSC17bWeak[which(crimeSC17bWeak$NUM_CRIMES <= 953), ]
crimeSC18cWeak <- crimeSC18bWeak[which(crimeSC18bWeak$NUM_CRIMES <= 953), ]
crimeSC19cWeak <- crimeSC19bWeak[which(crimeSC19bWeak$NUM_CRIMES <= 953), ]
```

```
ggplot() +
  geom_point(data=crimeSC15cWeak, aes(dayofyear, NUM_CRIMES, color="2015")) +
  geom_point(data=crimeSC16cWeak, aes(dayofyear, NUM_CRIMES, color="2016")) +
  geom_point(data=crimeSC17cWeak, aes(dayofyear, NUM_CRIMES, color="2017")) +
  geom_point(data=crimeSC18cWeak, aes(dayofyear, NUM_CRIMES, color="2018")) +
  geom_point(data=crimeSC19cWeak, aes(dayofyear, NUM_CRIMES, color="2019")) 
```
